# preparation

```python
import pandas as pd
import networkx as nx

# ======================================
# PARAMETERS
# ======================================
INPUT_FILE = "soc-twitter-higgs.csv"     # ملفك الأصلي
OUTPUT_FILE = "twitter_higgs_TOP2.csv"    # ملف الخرج
TOP_N = 50000     # غيّرها إلى 50000 إذا أردت

print("Loading dataset...")

# قراءة البيانات
df = pd.read_csv(INPUT_FILE)

print(f"Total edges: {len(df)}")

# بناء الرسم البياني الموجّه
G = nx.from_pandas_edgelist(
    df,
    source="Source",
    target="Target",
    create_using=nx.DiGraph()
)

print(f"Total nodes: {G.number_of_nodes()}")

# ======================================
# SELECT TOP N NODES BY DEGREE
# ======================================
print(f"Selecting top {TOP_N} nodes by degree...")

degree_dict = dict(G.degree())
top_nodes = sorted(
    degree_dict.items(),
    key=lambda x: x[1],
    reverse=True
)[:TOP_N]

top_nodes_set = set(node for node, _ in top_nodes)

# ======================================
# FILTER EDGES
# ======================================
print("Filtering edges...")

filtered_df = df[
    (df["Source"].isin(top_nodes_set)) &
    (df["Target"].isin(top_nodes_set))
]

print(f"Filtered edges: {len(filtered_df)}")

# ======================================
# SAVE TO CSV
# ======================================
filtered_df.to_csv(OUTPUT_FILE, index=False)

print("======================================")
print("DONE!")
print(f"Saved file: {OUTPUT_FILE}")
print(f"Nodes kept: {TOP_N}")
print("======================================")
```
